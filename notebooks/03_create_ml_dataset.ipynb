```python

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../processed_data/spotify_cleaned.csv")
print(df.shape)
```

```python
# =========================================================
# CREATE ML-READY DATASET
# =========================================================

ml_df = df.copy()
```

```python

# ---------------------------------------------------------
# DROP NON-ML COLUMNS
# ---------------------------------------------------------

DROP_COLUMNS = [
    "id",
    "name",
    "album_name",
    "artists",
    "artist_ids",
    "niche_genres",
    "year",
    "total_artist_followers"

]

ml_df = ml_df.drop(columns=DROP_COLUMNS)

print("After dropping columns:")
print(ml_df.shape)
```

```python

# ---------------------------------------------------------
# CONVERT BOOLEAN GENRES -> 0/1
# ---------------------------------------------------------

genre_cols = [
    col for col in ml_df.columns
    if col.startswith("genre_")
]

ml_df[genre_cols] = ml_df[genre_cols].astype(int)

print("\nConverted genre columns to binary.")
```

```python
# ---------------------------------------------------------
# ONE-HOT ENCODE ERA
# ---------------------------------------------------------

ml_df = pd.get_dummies(
    ml_df,
    columns=["era"],
    prefix="era"
)

# Convert era booleans too
era_cols = [
    col for col in ml_df.columns
    if col.startswith("era_")
]

ml_df[era_cols] = ml_df[era_cols].astype(int)

print("\nAfter era encoding:")
print(ml_df.shape)
```

```python
# ---------------------------------------------------------
# FINAL CHECK
# ---------------------------------------------------------

print("\nColumns:")
print(ml_df.columns.tolist())

print("\nData types:")
print(ml_df.dtypes)
```

```python
# ---------------------------------------------------------
# SAVE ML-READY DATASET
# ---------------------------------------------------------

OUTPUT_PATH = "../processed_data/spotify_ml_ready.csv"

ml_df.to_csv(OUTPUT_PATH, index=False)

print(f"\nSaved ML-ready dataset to:\n{OUTPUT_PATH}")
```
